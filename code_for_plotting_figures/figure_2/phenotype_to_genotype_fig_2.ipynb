```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

classes_original = ['DAPT', 'WT', 'her1;her7', 'tbx6']
classes_shuffled = ['WT', 'tbx6', 'DAPT', 'her1;her7']

def plot_distribution(matrix, xlabel, ylabel, title, shuffle):
    fig, ax = plt.subplots(figsize=(7.5, 7.5))
    ax.matshow(matrix, cmap=plt.cm.Blues, alpha=0.3)
    for i in range(matrix.shape[0]):
        for j in range(matrix.shape[1]):
            ax.text(x=j, y=i,s="{val:.2f}".format(val = matrix[i, j]), va='center', ha='center', size='xx-large')
    if(shuffle == 0):
        plt.xticks(np.arange(4), classes_original, fontsize = 15)
    if(shuffle == 1):
        plt.xticks(np.arange(4), classes_shuffled, fontsize = 15)
    #plt.yticks(np.arange(1),['Genotype class'], fontsize = 15)
    plt.yticks(np.arange(0),[''], fontsize = 15)
    plt.xlabel(xlabel, fontsize=18)
    plt.ylabel(ylabel, fontsize=18)
    plt.title(title, fontsize=18)
    plt.show()
    

classification_h1h7_homo_tbx6_homo = np.array([0, 6, 0, 1])/7
classification_h1h7_homo_tbx6_hetero = np.array([0, 0, 0, 6])/6
classification_h1h7_hetero_tbx6_homo = np.array([0, 12, 0, 2])/14
classification_wt = np.array([1, 0, 0, 0])/1 
shuffle = 1
plot_distribution(np.array([classification_h1h7_homo_tbx6_homo]), '', '', '', shuffle)
plot_distribution(np.array([classification_h1h7_homo_tbx6_hetero]), '', '', '', shuffle)
plot_distribution(np.array([classification_h1h7_hetero_tbx6_homo]), '', '', '', shuffle)
plot_distribution(np.array([classification_wt]), '', '', '', shuffle)
```
